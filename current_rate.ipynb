```
import time
import requests
import pandas as pd

url_symbols ="https://archives.nseindia.com/content/equities/EQUITY_L.csv"
r = requests.get(url_symbols)
with open("EQUITY_L.csv", 'wb') as f:
      f.write(r.content)
data =pd.read_csv("EQUITY_L.csv")
data=data['SYMBOL']
codelist=data.values.tolist()

endtime =int(time.time())
starttime=int(time.time())-86400*0  #one day back
for code in codelist:
  url="https://query1.finance.yahoo.com/v7/finance/download/{}.NS?period1={}&period2={}&interval=1d&events=history".format(code,starttime,endtime)
  r = requests.get(url)
  try:
    #print(r.content)
    print("{}: {}".format(code,str(r.content).split(",")[11]))
  except:
    print("{}: Ingnore".format(code))
```
